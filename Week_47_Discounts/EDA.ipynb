```python
import polars as pl
df = (
    pl.read_csv('Sales.csv')
    .select(
        DATE = pl.col('Order Date')
            .str.split(' ')
            .list.first()
            .str.to_date(format='%m/%d/%Y'),
        # ORDER = pl.col('Order ID'),
        # CUSTOMER = pl.col('Customer ID'),
        # LOCATION = pl.col('Location ID'),
        PRODUCT = pl.col('Product ID'),
        # QUANTITY = pl.col('Quantity').cast(pl.Int32),
        COST = pl.col('Sales').cast(pl.Float32),
        # DISCOUNT = pl.col('Discount').cast(pl.Float32),
        # SALES = pl.col('Sales').cast(pl.Float32),
        PROFIT = pl.col('Profit').cast(pl.Float32),
        COUNTRY = pl.col('Country').cast(pl.Categorical),
        SEGMENT = pl.col('Segment').cast(pl.Categorical),
    )
)
df
```

```python
import pycountry
import polars as pl
countries= [
    'Afghanistan', 'Argentina', 'Australia', 'Austria', 'Azerbaijan', 
    'Bangladesh', 'Brazil', 'Canada', 'China', 'Colombia', 'Denmark', 
    'Finland', 'France', 'India', 'Iran', 'New Zealand', 'Norway', 'Pakistan', 
    'Russian Federation', 'Saudi Arabia', 'South Africa', 'Sri Lanka', 'Switzerland', 
    'United Arab Emirates', 'United Kingdom', 'United States'
]
iso_codes = [pycountry.countries.lookup(c).alpha_3 for c in countries]
 
df_iso = (   # join this with main df to get ISO-3 codes for each country
    pl.DataFrame({
        'COUNTRY': countries,
        'ISO-3': iso_codes
    })
)
print(df_iso)
```
